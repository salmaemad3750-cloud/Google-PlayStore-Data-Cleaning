# Google Play Store Data Cleaning

## Objective
The objective of this project is to clean the Google Play Store dataset by:
- Identifying missing values
- Removing duplicate records
- Correcting data types
- Exporting a clean dataset for future analysis

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("googleplaystore.csv")

df.head()
```

```python
print("Shape:")
print(df.shape)

print("\nColumns:")
print(df.columns)

print("\nInformation:")
df.info()

print("\nSummary Statistics:")
display(df.describe(include="all"))
```

```python
print(df.isnull().sum())
```

```python
# Fill missing Rating values with the median
df["Rating"] = df["Rating"].fillna(df["Rating"].median())

# Fill missing categorical values with the mode
df["Type"] = df["Type"].fillna(df["Type"].mode()[0])

df["Content Rating"] = df["Content Rating"].fillna(
    df["Content Rating"].mode()[0]
)

df["Current Ver"] = df["Current Ver"].fillna(
    df["Current Ver"].mode()[0]
)

df["Android Ver"] = df["Android Ver"].fillna(
    df["Android Ver"].mode()[0]
)
```

```python
print(df.isnull().sum())
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df["Reviews"] = pd.to_numeric(df["Reviews"], errors="coerce")
```

```python
df["Installs"] = (
    df["Installs"]
    .str.replace(",", "", regex=False)
    .str.replace("+", "", regex=False)
)

df["Installs"] = pd.to_numeric(df["Installs"], errors="coerce")
```

```python
df["Price"] = (
    df["Price"]
    .str.replace("$", "", regex=False)
)

df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
```

```python
df["Size"] = df["Size"].replace("Varies with device", np.nan)

df["Size"] = (
    df["Size"]
    .str.replace("M", "", regex=False)
    .str.replace("k", "", regex=False)
)

df["Size"] = pd.to_numeric(df["Size"], errors="coerce")
# Fill missing values created during conversion
df["Size"] = df["Size"].fillna(df["Size"].median())
```

```python
df["Last Updated"] = pd.to_datetime(df["Last Updated"])
```

```python
df.to_csv("googleplaystore_cleaned.csv", index=False)
```
